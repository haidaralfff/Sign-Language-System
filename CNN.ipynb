```python
import numpy as np
import os
import cv2
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

data = []
labels = []

base_path = "dataset/SIBI"
img_size = 64

for folder in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder)

    if not os.path.isdir(folder_path):
        continue

    count = 0  # jumlah data yang sudah diambil dari folder dataset

    for file in os.listdir(folder_path):
        if file.endswith(".jpg") or file.endswith(".png"):

            if count >= 5:   # 馃敟 BATAS 5 DATA
                break

            img_path = os.path.join(folder_path, file)

            img = cv2.imread(img_path)
            img = cv2.resize(img, (img_size, img_size))
            img = img / 255.0

            data.append(img)
            labels.append(folder)

            count += 1

# encode label
unique_labels = sorted(list(set(labels)))
label_dict = {label: i for i, label in enumerate(unique_labels)}
y = np.array([label_dict[label] for label in labels])

X = np.array(data)
y = to_categorical(y)

print("Jumlah data total:", len(X))
print("Jumlah kelas:", len(unique_labels))

# MODEL
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(img_size, img_size, 3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(len(unique_labels), activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X, y, epochs=20)

accuracy = history.history['accuracy'][-1] * 100
print(f"Akurasi training: {accuracy:.2f}%")

os.makedirs("model", exist_ok=True)
model.save("model/cnn_model.h5")

print("Model berhasil dibuat!")
```
